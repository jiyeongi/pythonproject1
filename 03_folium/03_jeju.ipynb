```python
# 주소정보를 이용하여 위도, 경도 값을 저장합니다. '위치'컬럼에(위도,경도) 형태로 저장
#  주소를 찾아서 저장하는 함수를 작성하여 처리 
# 함수 이름은 geo_coder(query) 리턴값:(위도,경도)만일 못찾았다면 '결과없음'으로 리턴
# 함수 사용 apply()를 이용하여 데이터 프레임에 적용하기
```

```python

import requests

def geo_coder(query):
    REST_API_KEY='2dc6ae7b445f82a65a03a3ff81c941ea'
    code='PO3'
    query=query.split('(')[0]
    url=f'https://dapi.kakao.com/v2/local/search/keyword.json?query={query}&category_group_code={code}'
    header= {"Authorization": f"KakaoAK {REST_API_KEY}"}
    r = requests.get(url, headers=header)
    jsondata=r.json()
    try:
        return jsondata['documents'][0]['y'],jsondata['documents'][0]['x']
    except:
        return '결과없음' 
    
df['위치']=df['주소'].apply(geo_coder)
    
#     document = res.json()["documents"][0]
#     if len(documents) == 0:
#         return '결과없음'
#     document = documents[0]
#     return document["y"], document["x"]

# df = pd.DataFrame({'address': ['']})

# df['geo'], df['coder'] = zip(*df['address'].apply(get_lat_lon))
```

```python
#'위치'컬럼의 값을 marker위치값으로,tooltip 으로 읍면동을 사용하여 지도에 표시(클러스터이용)
# 함수사용,apply()를 사용
```

```python
import folium
from folium.plugins import MarkerCluster

def display(data,map):
    print(data[0])
    if data[0]!= '결과없음':
        folium.Marker(data[0],tooltip=data[1]).add_to(map)
    
map=folium.Map((33.3616666,126.5291666),zoom_start=9)
map_c=MarkerCluster().add_to(map)

df[['위치','읍면동']].apply(display,map=map_c,axis=1)

map
    
```

```python
map=folium.Map((33.3616666,126.5291666),zoom_start=9)
map_c=MarkerCluster().add_to(map)
# iterrows 데이터프레임의 각행을 반복적으로 처리하기
for index,row in df.iterrows():
    # print(index,row)
    if row['위치']!='결과없음':
        folium.Marker(row['위치'],tooltip=row['읍면동']).add_to(map_c)
map
```

```python
# 일반현황 컬럼의 인구수만 분리하여 인구수라는 int형 컬럼을 생성
```

```python
import re

df['일반현황'].apply(lambda x:int(re.findall('[0-9]{4,}',x)[-1]))
```
