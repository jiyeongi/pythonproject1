```python
person=pd.read_csv('data/survey_person.csv')
site=pd.read_csv('data/survey_site.csv')
survey=pd.read_csv('data/survey_survey.csv')
visited=pd.read_csv('data/survey_visited.csv')
```

```python
visited
```

```python
v_subset=visited.loc[[0,2,6]]
v_subset
```

```python
# pd.merge
site.merge(v_subset,left_on='name',right_on='site')
```

```python
site.merge(visited,left_on='name',right_on='site')
```

```python
ps=person.merge(survey,left_on='ident',right_on='person')
```

```python
vs=visited.merge(survey,left_on='ident',right_on='taken')
```

```python
ps.head(1)
```

```python
vs.head(1)
```

```python
# 2값을 연결하기
ps.merge(vs,
         left_on=['ident','taken','quant','reading'],
         right_on=['person','ident','quant','reading'],how='outer') #outer join
```
