```python
# 1.폰트파일로 FontProperties()
import matplotlib.font_manager as fm

path='D2CODING.TTC'
fontprop=fm.FontProperties(fname=path,size=14)

data=np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이',fontproperties=fontprop)
plt.ylabel('주식가격',fontproperties=fontprop)
plt.xlabel('시간(분)',fontproperties=fontprop)
```

```python
# 2.rcParams[]실행
# 기본폰트확인하기 
import matplotlib


print('설정파일위치:',matplotlib.matplotlib_fname())
print('기본설정폰트:',plt.rcParams['font.family'])
print('기본 폰트 크기 :',plt.rcParams['font.size'])
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['font.size']=12
plt.rcParams['axes.unicode_minus']=False
print('기본설정폰트:',plt.rcParams['font.family'])
print('기본 폰트 크기 :',plt.rcParams['font.size'])
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
```

```python
# sns.distplot(tips['total_bill'])
ax=sns.histplot(data=tips,x='total_bill',bins=8,kde=True)
ax.set_title('히스토그램')
```

```python
sns.__version__
```

```python
# 이상값으로 포현 (이상값이란 패턴에 맞지않는값같은것)
sns.countplot(data=tips,x='day')
ax.set_title('count of days')
ax.set_xlabel('day')
ax.set_ylabel('빈도')
```

```python
sns.regplot(data=tips,x='total_bill',y='tip',fit_reg=False)
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind="hex")
```

```python
sns.kdeplot(data=tips,x='total_bill',y='tip',fill=True)
```

```python
sns.barplot(data=tips,x='time',y='total_bill',fill=True)
```

```python
sns.boxplot(data=tips,y='total_bill',x='sex')
```

```python
sns.violinplot(data=tips,y='total_bill',x='sex')
```

```python
sns.pairplot(tips)
```
