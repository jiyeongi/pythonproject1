```python
# melt를 이용해서 월별 데이터를 '기간'컬럼으로 가격은 '평당분양가격'으로 처리하여 first_df_melf로 저장
first_df_melt=first_df.melt(id_vars='지역',var_name='기간',value_name='평당분양가격')
first_df_melt.head(3)
```

```python
last_df.info()
```

```python
# 기간 컬럼을 연도,  월 컬럼으로 분리하기 
first_df_melt['연도']=first_df_melt['기간'].str.split('년').str[0].astype(int)

first_df_melt['월']=first_df_melt['기간'].str.split('년').str[1].str.replace('월','').astype(int)
```

```python
first_df_melt.head(3)
```

```python
first_df_melt.info()
```

```python
# last_df 전용면적에서 전체 인 데이터만 last_df_prepare로 저장
# 값을 연결하기 


last_df_prepare = last_df[last_df['전용면적'] == '전체'].copy()
```

```python
last_df_prepare.head(2)
```

```python
first_df_melt.head(2)
```

```python
# 지역명을 지역컬럼으로 변경
last_df_prepare.columns[['지역명', '연도', '월', '분양가격', '평당분양가격', '전용면적']]
```

```python
# inplace는 판다스에서 데이터프레임을 변경할 때 사용하는 파라미터 중 하나입니다. 
# 이 파라미터를 True로 설정하면 데이터프레임이 변경되며, 
# False로 설정하면 데이터프레임이 변경되지 않고 새로운 데이터프레임이 반환됩니다. 
# 하지만 inplace=True를 사용하면 원본 데이터프레임이 변경되기 때문에 주의해서 사용해야 합니다. 
# 이 파라미터를 사용하지 않고 새로운 데이터프레임을 반환받아서 사용하는 것이 좋습니다.
last_df_prepare.rename(columns={'지역명':'지역'},inplace=True)
```

```python
# concat으로 데이터합치기
# concat() 함수를 사용하여 데이터프레임을 합치는 코드입니다. concat() 함수는 두 개 이상의 데이터프레임을 연결할 때 사용합니다. 이때 연결 방법은 행 방향으로 연결할 것인지 열 방향으로 연결할 것인지를 지정할 수 있습니다.
# 코드에서는 first와 last 두 개의 데이터프레임을 concat() 함수를 사용하여 행 방향으로 합칩니다. 
# 결과적으로 새로운 데이터프레임인 df가 생성됩니다. 
# 이어서 reset_index() 함수를 사용하여 인덱스를 재설정합니다. 
# 이때 drop=True를 지정하여 기존 인덱스를 삭제하고 새로운 인덱스를 생성합니다.
first=first_df_melt[['지역','연도','월','평당분양가격']]
last=last_df_prepare[['지역','연도','월','평당분양가격']]
df=pd.concat([first,last])
df.reset_index(drop=True,inplace=True)
df
```
